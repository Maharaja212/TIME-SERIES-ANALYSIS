```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("E:/221501073/Car_sales.csv")
```

```python
df.dropna(inplace=True)

```

```python

plt.figure(figsize=(25, 6))
sns.barplot(data=df, x='Manufacturer', y='Sales_in_thousands')

plt.xlabel('Manufacturer')
plt.ylabel('Sales_in_thousands')
plt.title('Maximum Sales of Each Company')

plt.legend(title='Company')
plt.show()
```
