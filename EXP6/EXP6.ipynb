```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Load the dataset from the sample data space in Colab
file_path = '/content/sample_data/Car_sales.csv'
df = pd.read_csv(file_path)
```

```python
df['Latest_Launch'] = pd.to_datetime(df['Latest_Launch'], errors='coerce')
df = df.sort_values(by='Latest_Launch').reset_index(drop=True)
```

```python
df['Time'] = np.arange(len(df))
```

```python
time = df['Time']
values = df['Sales_in_thousands']
```

```python
window_size = 5
df['Moving_Avg'] = values.rolling(window=window_size, center=True).mean()
```

```python
plt.figure(figsize=(25, 6))
plt.plot(time, values, label='Original Data', linestyle='dashed', alpha=0.7)
plt.plot(time, df['Moving_Avg'], label=f'{window_size}-Day Moving Average', color='red')
plt.xlabel('Time')
plt.ylabel('Car Sales (in thousands)')
plt.legend()
plt.title('Moving Average Smoothing of Car Sales Data')
plt.show()
```
